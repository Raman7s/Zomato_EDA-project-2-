### Adding Dataset Which Includes Its Country Code 

```python
zomato_data1 =pd.read_excel("Country-Code.xlsx")
zomato_data1
```

##### Now merging both the datasets

```python
zomato_final_data = pd.merge(zomato_data,zomato_data1,on ='Country Code', how= 'left')
```

```python
zomato_final_data.head(2)
```

```python
zomato_final_data.shape
```

```python
zomato_final_data.columns
```

```python
# checking the null values after merging both datasets

zomato_final_data.isnull().sum()
```

```python
zomato_final_data.info()
```

```python
zomato_final_data.describe()
```

```python
# Let's check total country name

country_name = zomato_final_data.Country.value_counts().index
country_name
```

```python
country_values = zomato_final_data.Country.value_counts().values
country_values
```

```python
# to show the top 5 country who use zomato's using pie chart
# pie chart
plt.figure(figsize=(12,15))
plt.pie(country_values[:5],labels = country_name[:5],autopct ='%1f%%')
```

#### Observation:-
           1. Plotted a pie chart to show the top 5 countries records or transaction in zomato.
           2. the first country is India followed by United State then United kingdom and the last two have same percentage    
              they are Brazil and UAE

```python
zomato_final_data.columns
```

```python
# Checking ratings and their counts

ratings = zomato_final_data.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating count'})
ratings
```

### Observation:
     1. Rating 0 shows in white color with the max count indicates----> Not Rated
     2. Ratings from 1.8 to 2.4 shows in red color which indicates---> POOR
     3. Ratings from 2.5 to 3.4 shows in orange which indicates---> Average
     4. Ratings from 3.5 to 3.9 shows in yellow which indicates---> Good
     5. Ratings from 4.0 to 4.4 shows in green which indicates---> Very Good
     6. Ratings from 4.5 to 4.9 shows in dark green which indicates---> Excellent 

##### Data Visualization 

```python
# Drawing some visualization according to ratings

ratings.head()
```

```python
# Bar plot showing the rating counts

plt.figure(figsize=(12,10))
sns.barplot(x='Aggregate rating',y='Rating count',data=ratings)
```

```python
# According to their rating counts their colors are mentioned

plt.figure(figsize =(10,9))
sns.barplot(x='Aggregate rating',y='Rating count',hue ='Rating color',data=ratings,palette =['blue','red','orange','yellow','green','green'])
```

#### Observation:
         1. Maximum count is 0.
         

```python
# count plot

sns.countplot(x='Rating color',data=ratings,palette =['blue','red','orange','yellow','green','green'])
```

```python
# Analysing zero rating given by the countries

zomato_final_data[zomato_final_data["Rating color"]=='White'].groupby('Country').size().reset_index()
```
